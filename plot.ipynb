```python
import matplotlib.pyplot as plt

# Define the coefficient values and corresponding compression ratios and compression ratios without trailing zeroes
coefficients = [1, 3, 6, 10, 15, 28]
compression_ratio_trailing = [7.27, 7.27, 7.27, 7.27, 7.27, 7.27]
compression_ratio_no_trailing = [660.55, 386.55, 256.83, 181.37, 151.03, 115.02]

# Define the previous RMSE and PSNR values
rmse_values = [5.44, 5.27, 5.19, 5.13, 5.09, 5.03]
psnr_values = [33.42, 33.69, 33.8, 33.93, 34.00, 34.10]

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the previous RMSE and PSNR values against the number of coefficients with dots and a triangle marker
ax.plot(coefficients, rmse_values, 'o-', label='Previous RMSE')
ax.plot(coefficients, psnr_values, '^-', label='Previous PSNR')

# Plot the compression ratio with trailing zeroes and without trailing zeroes against the number of coefficients with a line
ax.plot(coefficients, compression_ratio_trailing, '--', label='Compression Ratio (with trailing zeroes)')
ax.plot(coefficients, compression_ratio_no_trailing, '--', label='Compression Ratio (without trailing zeroes)')

# Set the axis labels and title
ax.set_xlabel('Number of coefficients')
ax.set_ylabel('Value')
ax.set_title('Compression Performance')

# Add a legend to the plot
ax.legend()

# Add a text box with the average PSNR value
avg_psnr = sum(psnr_values)/len(psnr_values)
ax.text(0.8, 0.2, f'Average PSNR: {avg_psnr:.2f}', transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='black', alpha=0.7))

# Customize the grid appearance
ax.grid(True, linestyle='--', alpha=0.5, axis='y')

# Add a background color to the plot
ax.set_facecolor('#f7f7f7')

# Display the plot
plt.show()
```
